```python
X = data.drop(('Legendary'), axis = 1) # выбрасываем столбец 'class'
y = data['Legendary']
feature_names = X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

# $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

### Подбор параметров

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

### Подбор _гипер_параметров

Подберем _оптимальное_ значение параметра регуляризации `alpha`

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (20,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'train')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

Обучаем много сетей с разным параметром _alpha_ и строим график зависимости ошибки

При _alpha_ > 10 наблюдается underfitting. При _alpha_ < 10 ошибка тестовой выборки колеблется в районе одного значения. Также, при _alpha_ < 1 наблюдается переобучение

```python
min_train_err = np.min(train_err[:-8])
min_test_err = np.min(test_err[:-8])
print(min_train_err, min_test_err)
```

Находим минимум ошибок начиная с восьмой точки, так как раньше находится значение тестовой ошибки, но есть недообучение

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
alpha_opt = alpha_opt[0]
```

### Классификатор при найденном _оптимальном_ значении `alpha`:

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (20,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

```python
plt.plot(neurons_arr, train_err, 'b-o', label = 'train')
plt.plot(neurons_arr, test_err, 'r-o', label = 'test')
plt.title('Error vs. neurons')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
neurons_opt = neurons_arr[test_err == min_test_err]
print(neurons_opt)
```

## Random Forest ##

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000, oob_score=True)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

```python
trees_arr = np.concatenate((np.arange(0, 10, 1)[1:], np.arange(10, 70, 5)[1:]))
test_err = []
train_err = []
for tree in trees_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = tree, random_state = 42)
    rf.fit(X_train, y_train)

    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.title('Error vs. trees number')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

При количестве деревьев > 20 ошибка на тренировочной выборке - 0. Тестовая достигает минимума на 45 деревьях, дальше начинается небольшое переобучение


```python
trees_arr = np.arange(20, 100, 5)
test_err = []
train_err = []
oob_err = []
for tree in trees_arr:
    rf = ensemble.RandomForestClassifier(n_estimators = tree, oob_score = True, random_state = 42)
    rf.fit(X_train, y_train)

    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - rf.oob_score_)
```

```python
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.title('Error vs. trees number')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

## ExtraTreesClassifier ##

```python
ExtrRF = ensemble.ExtraTreesClassifier(n_estimators = 100, oob_score=True, bootstrap = True)
ExtrRF.fit(X_train, y_train)

err_train = np.mean(y_train != ExtrRF.predict(X_train))
err_test  = np.mean(y_test  != ExtrRF.predict(X_test))

print(err_train, err_test)
```

```python
trees_arr_add = np.arange(1, 20, 1)
test_err = []
train_err = []
oob_err = []
for tree in trees_arr_add:
    ExtrRF = ensemble.ExtraTreesClassifier(n_estimators = tree, bootstrap = True, random_state = 42)
    ExtrRF.fit(X_train, y_train)

    y_train_pred = ExtrRF.predict(X_train)
    y_test_pred = ExtrRF.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))

trees_arr = np.arange(20, 70, 5)
for tree in trees_arr:
    ExtrRF = ensemble.ExtraTreesClassifier(n_estimators = tree, oob_score = True, bootstrap = True, random_state = 42)
    ExtrRF.fit(X_train, y_train)

    y_train_pred = ExtrRF.predict(X_train)
    y_test_pred = ExtrRF.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - ExtrRF.oob_score_)
```

```python
plt.plot(np.concatenate((trees_arr_add, trees_arr)), train_err, 'b-o', label = 'train')
plt.plot(np.concatenate((trees_arr_add, trees_arr)), test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.title('Error vs. trees number')
plt.xlabel('trees')
plt.ylabel('error')
plt.legend()
pass
```

Тренировочная ошибка достигает нуля при количестве деревьев > 10. OOB ошибка близка к тестовой, но немного больше. Переобучение начинается с 50 деревьев
